```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
data1 = pd.read_excel('s1-groundtruth-plot.xls')
data2 = pd.read_excel('s2-groundtruth-plot.xls')
data3 = pd.read_excel('s3-groundtruth-plot.xls')
data4 = pd.read_excel('s4-groundtruth-plot.xls')

data1 = data1.drop('Centroids', axis=1)
data2 = data2.drop('Centroids', axis=1)
data3 = data3.drop('Centroids', axis=1)
data4 = data4.drop('Centroids', axis=1)
```

```python
data1 = data1.iloc[14:]
data2 = data2.iloc[14:]
data3 = data3.iloc[14:]
data4 = data4.iloc[14:]

print(len(data1))
print(len(data2))
print(len(data3))
print(len(data4))
```

```python
# plt.scatter(data1.values.T[0],data1.values.T[1],s=2)
plt.scatter(data2.values.T[0],data2.values.T[1],s=2)
# plt.scatter(data3.values.T[0],data3.values.T[1],s=2)
# plt.scatter(data4.values.T[0],data4.values.T[1],s=2)

plt.show()
```

```python
mms = preprocessing.MinMaxScaler().fit(data2)
data2 = mms.transform(data2)
```

```python
kmeans = KMeans(n_clusters=15, random_state=0, max_iter=1)

kmeans.fit(data2)
```

```python
xx = data2[kmeans.labels_==0]
xx = xx.T
plt.scatter(xx[1], xx[0])
xx = data2[kmeans.labels_==1]
xx = xx.T
plt.scatter(xx[1], xx[0])
plt.show()
```

```python
for i in range(15):
    xx =data2[kmeans.labels_==i]
    xx = xx.T
    plt.scatter(xx[1], xx[0])

plt.show()
```
